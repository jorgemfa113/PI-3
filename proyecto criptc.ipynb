Se realizo concexion de las REST API  de FTX, siguiendo la documentacion 

https://docs.ftx.com/#overview

Primero importamos todos los mercado para conocer las monedas disponibles y sus caracterisaticas

```python
import pandas as pd
import requests

markets = requests.get('https://ftx.com/api/markets').json()
df = pd.DataFrame(markets['result'])
df.set_index('name', inplace = True)
df.T
```

A vontinuacion realizamos la busqueda de los precios historicos de cada una de las 10 monedas seleccionadas, a continuacion se muestra un ejemplo con la moneda CEL

```python
historical = requests.get('https://ftx.com/api/markets/CEL/USD/candles?resolution=86400&limit=100000').json()
historical = pd.DataFrame(historical['result'])
#historical.drop(['startTime'], axis = 1, inplace=True)
historical
```

Las parametros utilizados para realizar las consultas son 


* market_name= Nombre de la moneda eje BTC
* resolution= Nos indica el intervalo de tiempo de resultados esperados
* start_time= Nos indica la fecha de inicio de resultados en segundos
* end_time= Nos indica la fecha final de resultados en segundos
* limit= nos indica limite de resultados esperados

GET /markets/{market_name}/candles?resolution={resolution}&start_time={start_time}&end_time={end_time}
